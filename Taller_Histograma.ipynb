<h1 style="font-size:200%; background-color:powderblue; color:red; text-align:center; line-height : 60px;"> Cargar y visualizar las fotografias</h1>

<ul style="font-size:150%; line-height : 35px;">

<p style="text-align:justify">Con ayuda de la función de la libreria cv2, <span style="color:blue"><b>[cv2.imread()] </b></span> se pueden guardar las tres imagenes como si fueran matrices de dos dimensiones nxm, donde n es el alto y m el ancho de la imagen medido en pixeles. Como son leidas en escala de grises solo se puede acceder al dato correspondiente a la intensidad por cada pixel. </p>

</ul>

```python
#Cargar Imagenes
#asigna la imagen con iluminación interior en la variable img1
img1 = cv2.imread('Fig/Interior.jpg',0)
#asigna la imagen con iluminación sombra exterior en la variable img2
img2 = cv2.imread('Fig/Ext_Sombra.jpg',0)
#asigna la imagen con iluminación luz exterior en la variable img3
img3 = cv2.imread('Fig/Ext_Sol.jpg',0)


plt.figure(1)
#visualización en escala de grises la variable img1
plt.subplot(131), plt.imshow(img1,cmap=plt.cm.gray), plt.title('Interior')
#visualización en escala de grises la variable img2
plt.subplot(132), plt.imshow(img2,cmap=plt.cm.gray), plt.title('Exterior con Sombra')
#visualización en escala de grises la variable img3
plt.subplot(133), plt.imshow(img3,cmap=plt.cm.gray), plt.title('Exterior con Sol')
plt.subplots_adjust(top=2, bottom=0.2, left=0.10, right=2, hspace=0.3, wspace=0.5)
plt.show()
```

<h1 style="font-size:200%; background-color:powderblue; color:red; text-align:center; line-height : 60px;">
Histogramas de las fotografias </h1>


<ul style="font-size:150%; line-height : 35px;">

<p style="text-align:justify">Se, procede a ver los histogramas correspondientes a las tres imagenes.hay que recordar que el  histograma  es  una  función  de  densidad  de  probabilidad  deniveles  de  gris,  la  cual  relaciona  las  abscisas  operando  con  elrango  posible  de  intensidades  de  gris,  y  las  ordenadasoperando  con  el  número  de  píxeles  para  cada  nivel  de  gris. </p>

<p style="text-align:justify">Se debe hacer uso de la libreria cv2, y la función <span style="color:blue"><b> [cv2.calcHist()]</b></span> el cual toma como parametros la imagen cargado, o en este caso la matriz de pixeles con su intensidad de gris, el canales de visualización,la mascara, el tamaño del histograma y el rango de intensidad de valores que tendra el histograma.Se calculan los histogramas con variaciones de tamaño, con 256,32,16,8 y 2 bins, y asi observar varios umbrales del histograma. </p>

</ul>

```python
#HISTOGRAMAS
#Imagen 1
#se genera el histograma de la imagen 1 (iluminación interior) con:
#256 valores
hist1_256 = cv2.calcHist([img1],[0],None,[256],[0,256])
#32 valores
hist1_32 = cv2.calcHist([img1],[0],None,[32],[0,256])
#16 valores
hist1_16 = cv2.calcHist([img1],[0],None,[16],[0,256])
#8 valores
hist1_8 = cv2.calcHist([img1],[0],None,[8],[0,256])
#2valores
hist1_2 = cv2.calcHist([img1],[0],None,[2],[0,256])

#Imagen 2
#se genera el histograma de la imagen 2 (iluminación sombra exterior) con:
#256 valores
hist2_256 = cv2.calcHist([img2],[0],None,[256],[0,256])
#32 valores
hist2_32 = cv2.calcHist([img2],[0],None,[32],[0,256])
#16 valores
hist2_16 = cv2.calcHist([img2],[0],None,[16],[0,256])
#8 valores
hist2_8 = cv2.calcHist([img2],[0],None,[8],[0,256])
#2 valores
hist2_2 = cv2.calcHist([img2],[0],None,[2],[0,256])

#Imagen 3
#se genera el histograma de la imagen 3 (iluminación luz exterior) con:
#256 valores
hist3_256 = cv2.calcHist([img3],[0],None,[256],[0,256])
#32 valores
hist3_32 = cv2.calcHist([img3],[0],None,[32],[0,256])
#16 valores
hist3_16 = cv2.calcHist([img3],[0],None,[16],[0,256])
#8 valores
hist3_8 = cv2.calcHist([img3],[0],None,[8],[0,256])
#2 valores
hist3_2 = cv2.calcHist([img3],[0],None,[2],[0,256])
```

</ul>

<h1 style="font-size:200%; background-color:powderblue; color:red; text-align:center; line-height : 60px;">
Comparación de los histogramas </h1>

<ul style="font-size:150%; line-height : 35px;">

<p style = "text-align:justify">
La función <span style="color:blue"><b> [cv2.compareHist(hist1,hist2, int  method)] </b></span> tiene como parametros los histogramas a comparar, los histogramas hist1_256,hist2_256,hist3_256, que corresponden a los de tamaño 256bins, que son arreglos unidimensionales, asi que la funcion de comparacion comparara elemento por elemento de cada histogrma y dara como resultado un valor asociado a una caracteristica, el otro parametro es el metodo de comparacion que tomara el criterio para dar el resultado a la caracteristica. 
</p>

<p style = "text-align:justify">
Los metodos de comparación son: 
</p>

<p>
<li style = "text-align:justify">
<b> Correlación: </b> Es  un  índice  estadístico  que  mide  la  relación  lineal  entre dos  variables  cuantitativas,  su  valor  varía  en  el  intervalo de -1 a 1, en donde el valor resultante es -1 existe una divergencia maxima, si el valor es 0 no hay correlación y si da 1 es que hay coincidencia perfecta.
</li>
<li style = "text-align:justify">
<b> Chi-cuadrado: </b> Es un indice de comparación que muestra el nivel de diferencia que pueden tener los histogramas, si el indice muestra niveles bajos existe similitud entre las imagenes, pero si es cero, existe una combinacion perfecta, pero si existe diferencias, el indice puede llegar a ser muy elevado.
</li>
<li style = "text-align:justify">
<b> Intersección: </b> Los valores que calculan son el minimo entre los histogramas y suma los valores correspondientes al minimo, asi que muestran la similitud, si los indices son muy altos, entonces existe mayor similitud, pero si son bajos no, una similitud perfecta da un indice de 1 y una diferencia total da un desajuste total de 0.
</li>
<li style = "text-align:justify">
<b> Distancia  de  Bhattacharyya: </b> Los valores bajos indican, alta similitud,y los altos mala similitud, una pareja perfecta da un inidice de 0, y un desajuste total da 1.
</li>
</p>

</ul>

```python
# COMPARACIÓN DE HISTOGRAMAS 
comp = np.zeros((5,12))
# CORRELACIÓN 
comp[0,0] = cv2.compareHist(hist1_256,hist2_256,method = cv2.HISTCMP_CORREL)
comp[0,1] = cv2.compareHist(hist1_256,hist3_256,method = cv2.HISTCMP_CORREL)
comp[0,2] = cv2.compareHist(hist2_256,hist3_256,method = cv2.HISTCMP_CORREL)
comp[1,0] = cv2.compareHist(hist1_32,hist2_32,method = cv2.HISTCMP_CORREL)
comp[1,1] = cv2.compareHist(hist1_32,hist3_32,method = cv2.HISTCMP_CORREL)
comp[1,2] = cv2.compareHist(hist2_32,hist3_32,method = cv2.HISTCMP_CORREL)
comp[2,0] = cv2.compareHist(hist1_16,hist2_16,method = cv2.HISTCMP_CORREL)
comp[2,1] = cv2.compareHist(hist1_16,hist3_16,method = cv2.HISTCMP_CORREL)
comp[2,2] = cv2.compareHist(hist2_16,hist3_16,method = cv2.HISTCMP_CORREL)
comp[3,0] = cv2.compareHist(hist1_8,hist2_8,method = cv2.HISTCMP_CORREL)
comp[3,1] = cv2.compareHist(hist1_8,hist3_8,method = cv2.HISTCMP_CORREL)
comp[3,2] = cv2.compareHist(hist2_8,hist3_8,method = cv2.HISTCMP_CORREL)
comp[4,0] = cv2.compareHist(hist1_2,hist2_2,method = cv2.HISTCMP_CORREL)
comp[4,1] = cv2.compareHist(hist1_2,hist3_2,method = cv2.HISTCMP_CORREL)
comp[4,2] = cv2.compareHist(hist2_2,hist3_2,method = cv2.HISTCMP_CORREL)
```

```python
# INTERSECCIÓN
comp[0,3] = cv2.compareHist(hist1_256,hist2_256,method = cv2.HISTCMP_INTERSECT)
comp[0,4] = cv2.compareHist(hist1_256,hist3_256,method = cv2.HISTCMP_INTERSECT)
comp[0,5] = cv2.compareHist(hist2_256,hist3_256,method = cv2.HISTCMP_INTERSECT)
comp[1,3] = cv2.compareHist(hist1_32,hist2_32,method = cv2.HISTCMP_INTERSECT)
comp[1,4] = cv2.compareHist(hist1_32,hist3_32,method = cv2.HISTCMP_INTERSECT)
comp[1,5] = cv2.compareHist(hist2_32,hist3_32,method = cv2.HISTCMP_INTERSECT)
comp[2,3] = cv2.compareHist(hist1_16,hist2_16,method = cv2.HISTCMP_INTERSECT)
comp[2,4] = cv2.compareHist(hist1_16,hist3_16,method = cv2.HISTCMP_INTERSECT)
comp[2,5] = cv2.compareHist(hist2_16,hist3_16,method = cv2.HISTCMP_INTERSECT)
comp[3,3] = cv2.compareHist(hist1_8,hist2_8,method = cv2.HISTCMP_INTERSECT)
comp[3,4] = cv2.compareHist(hist1_8,hist3_8,method = cv2.HISTCMP_INTERSECT)
comp[3,5] = cv2.compareHist(hist2_8,hist3_8,method = cv2.HISTCMP_INTERSECT)
comp[4,3] = cv2.compareHist(hist1_2,hist2_2,method = cv2.HISTCMP_INTERSECT)
comp[4,4] = cv2.compareHist(hist1_2,hist3_2,method = cv2.HISTCMP_INTERSECT)
comp[4,5] = cv2.compareHist(hist2_2,hist3_2,method = cv2.HISTCMP_INTERSECT)
```

```python
# CHI-CUADRADO
comp[0,6] = cv2.compareHist(hist1_256,hist2_256,method = cv2.HISTCMP_CHISQR)
comp[0,7] = cv2.compareHist(hist1_256,hist3_256,method = cv2.HISTCMP_CHISQR)
comp[0,8] = cv2.compareHist(hist2_256,hist3_256,method = cv2.HISTCMP_CHISQR)
comp[1,6] = cv2.compareHist(hist1_32,hist2_32,method = cv2.HISTCMP_CHISQR)
comp[1,7] = cv2.compareHist(hist1_32,hist3_32,method = cv2.HISTCMP_CHISQR)
comp[1,8] = cv2.compareHist(hist2_32,hist3_32,method = cv2.HISTCMP_CHISQR)
comp[2,6] = cv2.compareHist(hist1_16,hist2_16,method = cv2.HISTCMP_CHISQR)
comp[2,7] = cv2.compareHist(hist1_16,hist3_16,method = cv2.HISTCMP_CHISQR)
comp[2,8] = cv2.compareHist(hist2_16,hist3_16,method = cv2.HISTCMP_CHISQR)
comp[3,6] = cv2.compareHist(hist1_8,hist2_8,method = cv2.HISTCMP_CHISQR)
comp[3,7] = cv2.compareHist(hist1_8,hist3_8,method = cv2.HISTCMP_CHISQR)
comp[3,8] = cv2.compareHist(hist2_8,hist3_8,method = cv2.HISTCMP_CHISQR)
comp[4,6] = cv2.compareHist(hist1_2,hist2_2,method = cv2.HISTCMP_CHISQR)
comp[4,7] = cv2.compareHist(hist1_2,hist3_2,method = cv2.HISTCMP_CHISQR)
comp[4,8] = cv2.compareHist(hist2_2,hist3_2,method = cv2.HISTCMP_CHISQR)
```

```python
# BHATTACHARYYA
comp[0,9] = cv2.compareHist(hist1_256,hist2_256,method = cv2.HISTCMP_BHATTACHARYYA)
comp[0,10] = cv2.compareHist(hist1_256,hist3_256,method = cv2.HISTCMP_BHATTACHARYYA)
comp[0,11] = cv2.compareHist(hist2_256,hist3_256,method = cv2.HISTCMP_BHATTACHARYYA)
comp[1,9] = cv2.compareHist(hist1_32,hist2_32,method = cv2.HISTCMP_BHATTACHARYYA)
comp[1,10] = cv2.compareHist(hist1_32,hist3_32,method = cv2.HISTCMP_BHATTACHARYYA)
comp[1,11] = cv2.compareHist(hist2_32,hist3_32,method = cv2.HISTCMP_BHATTACHARYYA)
comp[2,9] = cv2.compareHist(hist1_16,hist2_16,method = cv2.HISTCMP_BHATTACHARYYA)
comp[2,10] = cv2.compareHist(hist1_16,hist3_16,method = cv2.HISTCMP_BHATTACHARYYA)
comp[2,11] = cv2.compareHist(hist2_16,hist3_16,method = cv2.HISTCMP_BHATTACHARYYA)
comp[3,9] = cv2.compareHist(hist1_8,hist2_8,method = cv2.HISTCMP_BHATTACHARYYA)
comp[3,10] = cv2.compareHist(hist1_8,hist3_8,method = cv2.HISTCMP_BHATTACHARYYA)
comp[3,11] = cv2.compareHist(hist2_8,hist3_8,method = cv2.HISTCMP_BHATTACHARYYA)
comp[4,9] = cv2.compareHist(hist1_2,hist2_2,method = cv2.HISTCMP_BHATTACHARYYA)
comp[4,10] = cv2.compareHist(hist1_2,hist3_2,method = cv2.HISTCMP_BHATTACHARYYA)
comp[4,11] = cv2.compareHist(hist2_2,hist3_2,method = cv2.HISTCMP_BHATTACHARYYA)
```

```python
nmethod = ['Correlacion', 'Interseccion', 'Chi-Cuadrado', 'Bhattacharyya']
nhistog = ['h1 vs h2', 'h1 vs h3', 'h2 vs h3']
ncolumns = [(x, y) for x in nmethod for y in nhistog]
nindex =['256 Bins','32 Bins','16 Bins','8 Bins','2 Bins']
df1 = pd.DataFrame(data = comp,index=nindex,columns=ncolumns)
df1
```

<p style = "text-align:justify">
La tabla anterior evidencia los diferentes metodos de comparación utilizados entre los diferentes histogramas paracada uno de los niveles de Bins, se realizaron las comparaciones de los histogramas con el mismo número de Bins (como ejemplo el histograma 1 de 256 Bins, con el histograma 2 de 256 Bins).     
</p>    

<li style = "text-align:justify">
Con el metodo de correlación los valores se encuentran entre el intervalo de -1 a 1, al evidenciar los datos obtenidos se detalla que cuando solo se cuenta con dos niveles de Bins en el histograma no existe una correlación entre los histogramas 1 y 3 con el histograma 2, pero que entre el histograma 1 y 3 existe una conincidencia perfecta. Pero estos valores no son los adecuados para detrminar la comparación de los histogramas, por ello cuando se cuenta con una mayor numero de Bins se observa que los valores no se encuentran en los extremos del intervalo, dichos valores nunca son cercanos a -1 o a 1, extos valores estan muy cercanos entre si, sin considerar el numero de Bins, por lo que se podria determinar que no existe una correlación entre los histogramas.
</li>

<li style = "text-align:justify">
Con el metodo de intersección los valores arrojados son muy altos, por lo que segun la teoria se encuentra una buena similitud entre ellos. De nuevo un histograma con dos niveles de bins no es adecuado para determinar la similitud de los histogramas, en los otros casos los valores son parecidos pero auemntan a medida que se disminuye los niveles en el histograma. Con los valores obtenidos se puede determinar que la mejor relación se encuentra en la fotografia del objeto con iluminación exterior y la fotografia de iluminación interior.    
</li>

<li style = "text-align:justify">
Al utilizar el metodo Chi-cuadrado se evidencia que los valores obtenidos son muy elevados, por lo cual se puede decir que no existe similitud entre los histogramas analizados. Los menores valores obtenidos se encuntran en la compración entre los histogramas 1 y 3 sin considerar el numero de Bins.
</li>

<li style = "text-align:justify">
Con la distancia Bhattacharyya el rango esta entre 0 y 1, en ninguno de los casos se encuentra el valor de 1, que indicaria una total diferencia entre los histogramas, pero tampoco existen casos donde el valor obtenido sea cero que evidenciaria una similitud perfecta. Los valores mas bajos se encuentran entre los histogramas 1 con los histogramas 2 y 3. Los valores de comparación entre los histogramas de 2 bins de nuevo no permiten determinar una correcta comparación entre los histogramas ya que la variación con respecto a los histogramas de mayores Bins es muy notable.
